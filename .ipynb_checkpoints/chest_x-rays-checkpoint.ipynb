```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline
from fastai import *
from fastai.vision import *
import os
```

```python
os.getcwd()
```

```python
path='./chest_x-rays_dataset/'
```

```python
data=ImageDataBunch.from_folder(path,valid='val',ds_tfms=get_transforms(do_flip=True),size=224,bs=8)
data.normalize(imagenet_stats)
```

```python
data.show_batch(rows=2,figsize=(7,6))
data.classes
data.c
```

```python
model=cnn_learner(data,models.resnet34,metrics=accuracy)
```

```python
model.fit_one_cycle(1)
```

```python
interpret=ClassificationInterpretation.from_learner(model)
```

```python
interpret.plot_confusion_matrix()
```

```python
test_data=ImageDataBunch.from_folder(path,test='test',size=224).normalize(imagenet_stats)
```

```python
test_data
```

```python
model.validate(test_data.test_dl)
```
